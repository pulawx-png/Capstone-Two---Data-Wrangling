```python
df = df.drop_duplicates()

df.duplicated().sum()
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
)

df.columns
```

```python
# Example: if you have a date/time column
# change 'date_time' to your actual column name
if 'date_time' in df.columns:
    df['date_time'] = pd.to_datetime(df['date_time'], errors='coerce')

# Example: convert a numeric column stored as string
# change 'traffic_volume' to your actual column name
if 'traffic_volume' in df.columns:
    df['traffic_volume'] = pd.to_numeric(df['traffic_volume'], errors='coerce')

df.dtypes
```

```python
# Example: fill numeric columns with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Example: fill object (string) columns with mode
obj_cols = df.select_dtypes(include=['object']).columns
for col in obj_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum()
```

```python
df.describe()
```

```python
# Example: remove rows where traffic_volume is above 99th percentile
if 'traffic_volume' in df.columns:
    upper = df['traffic_volume'].quantile(0.99)
    df = df[df['traffic_volume'] <= upper]

df.describe()
```

```python
# Example: filter rows where traffic_volume is greater than 4000
if 'traffic_volume' in df.columns:
    high_traffic = df.loc[df['traffic_volume'] > 4000]
    high_traffic.head()
```

```python
df.to_csv('traffic_cleaned.csv', index=False)
```
